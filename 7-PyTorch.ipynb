# Differentiation

```python
x = torch.linspace(-4,4,100,requires_grad=True)
Y = x**2
y = torch.sum(Y)
y.backward() #Calculates derivative
x.grad
plt.plot(x.detach().numpy(),Y.detach().numpy())
plt.plot(x.detach().numpy(),x.grad.detach().numpy())
```

```python
u=torch.tensor([1.,-2.,6.,4.],requires_grad=True)
v=torch.tensor([3.,-7.,0.,5.],requires_grad=True)
z=torch.exp(u)
w=v**3
Y=z+w
y=torch.sum(Y)
y.backward()
print(u.grad,v.grad)
```

```python
x = torch.linspace(-10,10,100,requires_grad=True)
Y = torch.relu(x)
y = torch.sum(Y)
y.backward() #Calculates derivative
x.grad
plt.plot(x.detach().numpy(),Y.detach().numpy())
plt.plot(x.detach().numpy(),x.grad.detach().numpy())
```

# Linear Regression

```python
X = torch.arange(-3,3,.1)
f = 1 * X - 1 
Y = f + .3 * torch.randn(X.size())
plt.plot(X.numpy(),Y.numpy(),linestyle = 'None',marker='+',color='r')
plt.plot(X.numpy(),f.numpy())
```

```python
def forward(x):
    return w*x+b
```

```python
def criterion(yhat,y):
    return torch.mean((yhat-y)**2)
```

```python
w = torch.tensor(10.,requires_grad=True)
b = torch.tensor(20.,requires_grad=True)
lr = .2
for epoch in range(100):
    Yhat = forward(X)
    loss = criterion(Yhat,Y)
    loss.backward()
    print(epoch,"w,b:",w.data,b.data)
    #print(w.grad.data,b.grad.data)
    w.data = w.data - lr*w.grad.data
    b.data = b.data - lr*b.grad.data
    
    w.grad.data.zero_()
    b.grad.data.zero_()
```
